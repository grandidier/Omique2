```python
def counting_sort(array,ind):
    list_index = [0]*len(array)
    for i in range(len(array)):
        list_index[i]=array[i][ind]
        
    x = max(list_index)
    count_arr= [0]*(x+1)
    sorted_arr=[0]*len(array)
    j=0
    
    for i in range(0, len(array)):
        index = array[i][ind]
        count_arr[index] += 1
        
    for i in range(1, len(count_arr)):
        count_arr[i] += count_arr[i - 1]
       
    i = len(array) - 1
    
    while i >= 0:
        index = array[i][ind]
        sorted_arr[count_arr[index] - 1] = array[i]
        count_arr[index] -= 1
        i -= 1
        #print("sort", sorted_arr)
        
    array = sorted_arr
    return array
```

```python
def RadixSort(array):
    x = len(array[0])
    ind=x-1
    sort=array
    while ind>=0 :
        sort=counting_sort(sort,ind)
        ind -=1
    return sort
```

```python
def index_merge(T,index0,index12):
    res=[]
    i=0
    j=0
    while i < len(index0) and j < len(index12):
        a=index0[i]
        b=index12[j]
        if T[a]<T[b]:
            res.append(a)
            i=i+1
        elif T[a]>T[b]:
            res.append(b)
            j=j+1
        else:
            if b%3==1:
                c = index12.index(a+1)
                d = index12.index(b+1)
                if c<d:
                    res.append(a)
                    i=i+1
                elif d<c:
                    res.append(b)
                    j=j+1
    
            if b%3==2:
                if T[a+1]<T[b+1]:
                    res.append(a)
                    i=i+1
                elif T[a+1]>T[b+1]:
                    res.append(b)
                    j=j+1
                if T[a+1]==T[b+1]:
                    c = index12.index(a+2)
                    d = index12.index(b+2)
                    if c<d:
                        res.append(a)
                        i=i+1
                    elif d<c:
                        res.append(b)
                        j=j+1
    if i == len(index0):
        for index in index12[j:]:
            res.append(index)
    elif j==len(index12):
        for index in index0[i:]:
            res.append(index)
    return res
                 
```

```python
def dc3(S):
    """
    Compute the suffix array
    
    Args:
        S (str): string
    
    Return:
        list of int: index list of suffix
    """
    #print("S : " + str(S))
    T=[]
    if type(S)==str:
        T = [ord(char) for char in S]
    else:
        T=S
    #print("T : " + str(T))
    T.extend((0,0,0))
    P0 = [i for i in range(len(T)-1) if ( i%3==0 and i+2<len(T))]
    P1 = [i for i in range(len(T)-1) if ( i%3==1 and i+2<len(T))]
    P2 = [i for i in range(len(T)-1) if ( i%3==2 and i+2<len(T))]
    P12 = P1+P2
    R12 = [(T[i],T[i+1],T[i+2]) for i in P12]
    R_sorted=RadixSort(R12)
    #print("P0 : " + str(P0))
    #print("P1 : " + str(P1))
    #print("P2 : " + str(P2))
    #print("R12 : " + str(R12))
    #print("R_sorted : " + str(R_sorted))
    
    duplicate=False
    order = [1]
    for i in range(1,len(R_sorted)):
        if R_sorted[i-1]==R_sorted[i]:
            order.append(order[-1])
            duplicate=True
        else:
            order.append(order[-1]+1)
    #print("order : " + str(order))    
    
    index=[]
    for elem in R_sorted:
        for i in P12:
            if (elem==(T[i],T[i+1],T[i+2]) and i not in index):
                index.append(i)
    #print("index : " + str(index))
    
    if duplicate:
        T_prime = []
        for elem in R12:
            found = False
            for i in range(len(order)):
                if elem==R_sorted[i] and found==False:
                    T_prime.append(order[i])
                    found=True
        index=[]
        index_prime = dc3(T_prime)
        for ind in index_prime:
            index.append(P12[ind])
            
    
        
  #Step2      
    R0 = []
    for num in P0:
        if num+1 in index:
            R0.append([T[num],index.index(num+1)])
    
    R0_sorted = RadixSort(R0)
    #print("R0_sorted : " + str(R0_sorted))
    
    index0=[]
    for elem in R0_sorted:
        index0.append(index[elem[1]]-1)

    
 #Step3   
    #print("T : " + str(T))
    #print("index0 : " + str(index0))
    #print("index : " + str(index))

    index_merged = index_merge(T,index0,index)  
    #print("index_merged : " + str(index_merged))
    
    #if 0 index_merged has no centinel
    new_index = index_merged
    for i in range(len(index_merged)):
        if T[index_merged[i]]==0:
            new_index=index_merged[:i]+index_merged[i+1:]
        
    #print("new_index : " + str(new_index))
        
    return new_index
```

```python
def get_BWT(S):
    index = dc3(S)
    res=[]
    for ind in index:
        res.append(S[ind-1])
    return res
```
